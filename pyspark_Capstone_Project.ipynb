```python
#Now I will convert this dataframe into a spark dataframe
#Initializing PySpark
from pyspark import SparkContext, SparkConf
from pyspark.sql import SparkSession
from pyspark.sql import SQLContext
# #Spark Config
conf = SparkConf().setAppName("sample_app")
sc = SparkContext(conf=conf)
spark = SparkSession.builder.appName('Test').getOrCreate()
sqlContext = SQLContext(sc)
```

```python
spark_df = spark.read.format("csv").option("header", "true").option("inferschema", "true").load("/content/olympics_new")
```

```python
spark_df.show()  #here I am using that cleaned dataset that I have analysed using EDA 
```

```python
# enableing the Apache Arrow for converting
# Pandas to pySpark DF(DataFrame)
spark.conf.set("spark.sql.execution.arrow.enabled", "true")
  
# Show the DataFrame
spark_df.show(10)
```

```python
spark_df.select("sports").distinct().count()  #counting the number of unique sports in the column
```

```python
# Now I will get the  the Olympic Sports/games in this dataset.
spark_df.select("sports").distinct().show(49)
```

```python
#Now I will write a PySpark code to plot the total number of medals in  each Olympic Sport/game
import pyspark.sql.functions as f 

total_medals = spark_df.groupBy("sports").sum("total_medal")
total_medals.orderBy(f.desc("sum(total_medal)")).show(49)
```

```python
Pdf = spark_df.toPandas ()
Pdf
```

```python
sports_sort1 = Pdf.groupby(Pdf['sports'], as_index=False, sort= True, group_keys=True)['total_medal'].sum()
   
sports_sort_1 = sports_sort1.sort_values(by = 'total_medal',ascending = True)
sports_sort_1 = sports_sort_1.iloc[:,:]
sports_sort_1
```

#In this way I got the sports wise medals so that I can visualize them as per the medals won by each sport.

```python
import plotly.express as px
fig = px.funnel(sports_sort_1, x='sports', y='total_medal')
fig.show()

```


*  Given is the plot for the total_medals won in each sport.

*  I plotted the funnel chart to plot the sports wise medals won

*   I found that highest Number of medals have been won in Swimming(765)

*   Least number of medals have been won in skeleton sports(18 only)

```python
# Now i will find out the total number of medals won by each country in swimming.
total_medalbycountry_swimming = spark_df.filter(spark_df.sports == 'Swimming').groupBy("country").sum("total_medal")
total_medalbycountry_swimming.orderBy(f.desc("sum(total_medal)")).show(110)
```

```python
# Let's Find the total number of medals won by each country in Skeleton
total_medalbycountry_skeleton = spark_df.filter(spark_df.sports == 'Skeleton').groupBy("country").sum("total_medal")
total_medalbycountry_skeleton.orderBy(f.desc("sum(total_medal)")).show(110)
```

```python
#USA is coming as a strong olympic contestent so let's find out the number of medals that the US won yearly
usa_medals = spark_df.filter(spark_df.country == 'United States').groupBy("year").sum("total_medal")
usa_medals.orderBy("sum(total_medal)").show()
```

```python
#Now I have to Find the total number of medals won by each country.
total_medals_c = spark_df.groupBy("country").sum("total_medal")
total_medals_c.orderBy(f.desc("sum(total_medal)")).show(110)
```
