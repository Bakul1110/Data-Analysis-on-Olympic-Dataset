```python
import numpy as np
import pandas as pd 
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv("/content/Iris.csv")
iris.head()
```

```python
iris_setosa = iris.loc[iris["Species"] == 'Iris-setosa']
counts, bin_edges = np.histogram(iris_setosa['PetalLengthCm'], bins = 10, density = True)
counts
```

```python
pdf = counts/(sum(counts))
pdf
```

```python
print(bin_edges)
```

```python
cdf = np.cumsum(pdf)
cdf
```

```python
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:],cdf)
plt.xlabel("PetalLength")
plt.ylabel("Probability")
plt.show()
```

# **The probability plot**
As we can see the plot is showing the probability of a setosa flower having the petallength on the x axis for example the probability of setosa flower having petallength 1.1 is 0.05 approx.

```python
#Histograms for each flower species
sns.FacetGrid(iris, hue = 'Species', height = 5).map(sns.histplot, 'PetalLengthCm', kde = True).add_legend();
plt.show()
```
